```python
# Separe a coluna 'duration' em 'duration_int' (numérico) e 'duration_unit'(unidade):


df_movies = df[df["type"] == "Movie"]

def takeFirstPart(x):
    return int(x.split(" ")[0])

def takeSecondtPart(x):
    return x.split(" ")[1]

df_movies["duration_int"] = df_movies["duration"].apply(takeFirstPart)
df_movies["duration_unit"] = df_movies["duration"].apply(takeSecondtPart)

df_movies.head(20)
```

```python
df_seasons = df[df["type"] == "TV Show"]

def takeFirstPart(x):
    return int(x.split(" ")[0])

df_seasons["seasons_int"] = df_seasons["duration"].apply(takeFirstPart)

df_seasons.head(20)
```

```python
max_seasons = df[df['type'] == 'TV Show'].loc[df[df['type'] == 'TV Show']['duration'].str.extract(r'(\d+)').astype(int).idxmax()]

max_movie_time = df[df['type'] == 'Movie'].loc[df[df['type'] == 'Movie']['duration'].str.extract(r'(\d+)').astype(int).idxmax()]

print(f"Série: {max_seasons['title']} - {max_seasons['duration']}")
print(f"Filme: {max_movie_time['title']} - {max_movie_time['duration']}")
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration_min'] = movies['duration'].str.extract(r'(\d+)').astype(int)

comedias = movies[movies['listed_in'].str.contains('Comedies', na=False)]['duration_min']
dramas = movies[movies['listed_in'].str.contains('Dramas', na=False)]['duration_min']

media_comedias = comedias.mean()
media_dramas = dramas.mean()

print(f"Duração média comédias: {media_comedias:.1f} minutos")
print(f"Duração média dramas: {media_dramas:.1f} minutos")
print(f"Quantidade comédias: {len(comedias)}")
print(f"Quantidade dramas: {len(dramas)}")
print(f"Hipótese é {'VERDADEIRA' if media_comedias < media_dramas else 'FALSA'}")
```

Com base nos dados obtidos, a hipótese foi testada através da **comparação das durações médias entre filmes de comédia e drama** no catálogo da Netflix. O código filtrou apenas filmes (excluindo as séries), extraiu a duração em minutos e separou os filmes por gênero baseado nas palavras "Comedies" e "Dramas" na coluna de categorias. Após calcular as médias de duração para cada grupo, a hipótese de que "filmes de comédia têm duração menor que filmes de drama" foi validada comparando se que a média de comédias é inferior à média de dramas, fornecendo uma resposta objetiva (VERDADEIRA ou FALSA) baseada nos dados reais do dataset.
